```python
import pandas as pd
from sklearn.model_selection import train_test_split
import tensorflow as tf
from tensorflow.keras.models import Sequential
from sklearn.metrics import accuracy_score, confusion_matrix
```

## Carregando Arquivos

```python
#Previsores
X = pd.read_csv("/content/entradas_breast.csv", sep=",")
X
```

```python
# Classe
y = pd.read_csv("/content/saidas_breast.csv")
y
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25) # 25% dos dados será para treino e 25% para validar os dados
```

```python
X_train.shape, y_train.shape
```

```python
X_test.shape, y_test.shape
```

```python
rede_neural = Sequential([
    tf.keras.layers.InputLayer(shape = (30,)),
    tf.keras.layers.Dense(units=16, activation="relu", kernel_initializer="random_uniform"),
    tf.keras.layers.Dense( units=1, activation="sigmoid")
])
```

```python
rede_neural.summary()
```

Configurando Parâmentros da Rede

```python
# Optimizer é o algoritmo que será utilizado para realizar a optimização dos pesos
# Loss é a função que fara o cáculo dos erros.
# Metrics lista com as métricas que serão retornadas.

rede_neural.compile(optimizer="adam", loss="binary_crossentropy", metrics=['binary_accuracy'])
```

```python
# treinamento da rede

rede_neural.fit(X_train, y_train, batch_size=10, epochs=100)
```

```python
previcoes = rede_neural.predict(X_test)
```

```python
previsoes = previcoes > 0.5
```

```python
previsoes
```

```python
y_test
```

```python
accuracy_score(y_test ,previsoes)
```

```python
confusion_matrix(y_test, previsoes)
```
